### 3) verifica condizionali

```python
%%pytest -v
import pytest

# modificare le seguenti funzioni per far eseguire i test con successo

def maggiore_di_uno(value) -> bool:
    # ritorna un valore booleano (true se il valore e' maggiore di uno, false se non lo e')
    #... code
    return value > 1


def minore_di_uno(value) -> bool:
    # ritorna un valore booleano (true se il valore e' minore di uno, false se lo e')
    #... code
    return value < 1


def carattere_in_stringa(char: str, search_string: str) -> bool:
    """
    verifica se un carattere Ã¨ contenuto  in una stringa. 
    param char: carattere da cercare
    param search_string: stringa in cui ricercare il carattere

    esempio: o Ã¨ contenuta in ciao, k non Ã¨ contenuta in prova
    """
    # ritorna un  valore booleano (true se il carattere(char) e' contenuto nella stringa(search_string) )
    return char in search_string


def numero_pari(number: int) -> bool:
    """
    verifica  se il numero passato come argomento Ã¨ pari
    """
    return number % 2 == 0


def numero_dispari(number: int) -> bool:
    """
    verifica  se il numero passato come argomento Ã¨ dispari
    """
    return not numero_pari(number)


def conteggio_dolci(persone: int, 
                    dolci_disponibili: int, 
                    numero_minimo_dolci: int = 3) -> bool:
    """ 
    la funzione ci  permette di verificare se i  dolci a nostra disposizione sono  sufficienti rispetto agli invitati
    ipotizzando che ogni  invitato  mangia  3 dolci.
    
    ##Â parameter ##
    param persone: numero di invitati alla festa
    param dolci_disponibili: numero  di dolci disponibili
    param numero_minimo_dolci: rappresenta il numero minimo di dolci che devono essere disponibili 
    
    es:
    10 persone , 30 dolci disponibili , 3 numero minimo  == true (i dolci sono sufficenti per la riuscita della festa)
    10 persone , 25 dolci disponibili , 3 numero minimo  == false (i dolci non bastano)
    
    """
    # la funzione deve verificare se i dolci disponibili sono sufficienti per 
    # le persone  invitate alla festa.
    # il numero minimo di dolci per ogni persona Ã¨ pari a 3 ma Ã¨ possibile personalizzarlo passando il parametro numero_minimo=N
    return ( dolci_disponibili / 3 ) >= persone


def ascensore_troppo_carico(persone: int, 
                            carico_consigliato: int = 1200, 
                            peso_medio: int = 80, 
                            extra: int = 10) -> bool:
    """
    param persone: numero di persone attualmente presenti  in ascensore
    param carico_consigliato: peso massimo in kg 
    param peso_medio: valore che  indica il peso medio di ogni persona
    param extra: percentuale di tolleranza di carico ulteriore rispetto al carico_consigliato
    
    es: 
    10 persone
    1200 carico consigliato
    80 ppeso medio
    10% tolleranza extra
    carico massimo sopportabile:  1200 + 10% (1320kg carico massimo)
    ascensore_sovraccarico? no (80kg per persona, 10 persone, max sopportabili 1320)
    
    calcolare se il peso totale  delle persone a bordo dell'ascensore sia  superiore al carico  massimo sopportabile
    es: 2000kg peso max, 2000kg attuali = False
    2000kg peso max, 1999kg attuali = True
    """
    peso_max = carico_consigliato  + ( carico_consigliato / 100 * extra)
    return (persone * peso_medio) >= peso_max
    
    
    
def benzina_sufficiente(km_per_litro: int, km_da_percorrere: int, litri_disponibili: int) -> bool:
    """
    param km_per_litro: i kilometri  che la macchina puÃ² percorrere con un litro di benzina
    param km_da_percorrere: la distanza da percorrere
    param litri_disponibili: litri di benzina disponibili nella macchina
    """
    return km_per_litro * litri_disponibili >= km_da_percorrere
    



################################################################################################
################################################################################################
#              TESTING - non toccare                                                           #
################################################################################################
@pytest.mark.parametrize("a,needle", [
    (1, False), (2, True), (0, False), (1.1, True)
])
def test_maggiore_di_uno(a, needle):
    assert maggiore_di_uno(a) == needle
    
@pytest.mark.parametrize("a,needle", [
    (1, False), (2, False), (0, True), (1.1, False)
])
def test_minore_di_uno(a, needle):
    assert minore_di_uno(a) == needle
    
@pytest.mark.parametrize("a,b,needle", [
    ("o", "ciao", True),("a", "prove", False),("k", "key", True),
])
def test_carattere_in_stringa(a, b, needle):
    assert carattere_in_stringa(a, b) == needle
    
@pytest.mark.parametrize("a,needle", [
    (1, False), (2, True), (10, True), (1.1, False)
])
def test_numero_pari(a, needle):
    assert numero_pari(a) == needle
    
@pytest.mark.parametrize("a,needle", [
    (1, True), (2, False), (10, False), (3, True)
])
def test_numero_dispari(a, needle):
    assert numero_dispari(a) == needle
    
@pytest.mark.parametrize("a,b,needle", [
    (10, 32, True), (15, 30, False), (7, 21, True)
])
def test_conteggio_dolci(a, b, needle):
    assert conteggio_dolci(a, b) == needle
    
@pytest.mark.parametrize("persone,carico,extra,needle", [
    (10, 800, 10, False), (13, 800, 10, True), (13, 1000, 5, False)
])
def test_ascensore_troppo_carico(persone, carico, extra, needle):
    assert ascensore_troppo_carico(persone=persone, carico_consigliato=carico, extra=extra) == needle
    
@pytest.mark.parametrize("a,b,c,needle", [
    (25, 800, 30, False), (25, 1000, 45, True), (25, 50, 50, True)
])
def test_benzina_sufficiente(a, b, c, needle):
    assert benzina_sufficiente(a, b, c) == needle
```
